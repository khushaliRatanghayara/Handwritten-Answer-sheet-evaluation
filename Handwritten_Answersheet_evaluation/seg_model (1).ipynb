```python
import imghdr

def is_image(filename):
    """
    Check if the given file is an image.
    Returns True if the file is an image, False otherwise.
    """
    image_type = imghdr.what(filename)
    return image_type is not None
```

```python
import pandas as pd
import cv2
import os

# Load DataFrame with bounding box data
df = pd.read_csv('/content/drive/MyDrive/seg.csv')

# Group DataFrame by image path
groups = df.groupby('image_path')

# Loop through each image path group
for image_path, group in groups:
    # Load image
    img = cv2.imread(image_path)

    # Create folder for image if it doesn't exist
    folder_name = os.path.splitext(os.path.basename(image_path))[0]
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    # Loop through bounding boxes and extract sub-image
    for i, row in group.iterrows():
        x, y, w, h = row['x'], row['y'], row['width'], row['height']
        sub_img = img[y:y+h, x:x+w]

        # Save sub-image to file in corresponding image folder
        sub_img_path = os.path.join(folder_name, f'sub_image_{i}.jpg')
        cv2.imwrite(sub_img_path, sub_img)
```

```python
from PIL import Image
from google.colab.patches import cv2_imshow
import os
import numpy as np
import cv2
import time

# get the list of image files in the folder
folder_path = "/content/drive/MyDrive/Output"
image_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg') or f.endswith('.jpeg') or f.endswith('.png')]

# loop through each image file
for file_name in image_files:
    # open the image
    image_path = os.path.join(folder_path, file_name)
    img = Image.open(image_path)

    # show the image
    cv2_imshow(np.array(img))

    # wait for 5 seconds
    time.sleep(5)

    # close the image window
    cv2.destroyAllWindows()

    # ask user if they want to rotate the image
    user_input = input("Do you want to rotate the image? (yes/no) ")

    if user_input.lower() == "yes":
        # ask for degree of rotation
        degree_input = int(input("Enter degree of rotation (90, 180, 270): "))

        # rotate the image
        rotated_img = img.rotate(degree_input, expand=True)

        # replace original image with rotated image
        rotated_img.save(image_path)
        print(f"{file_name} rotated by {degree_input} degrees and saved.")
    else:
        print(f"{file_name} not rotated.")
```
